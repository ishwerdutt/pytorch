## Converting numpy arrays to pytorch tensors

```python
X_train_tensor = t.from_numpy(X_train)
X_test_tensor = t.from_numpy(X_test)
y_train_tensor = t.from_numpy(y_train)
y_test_tensor = t.from_numpy(y_test)
```

# Define the model/Neural Network

```python
class cancer_classifier():
  def __init__(self, x):
    self.weights = t.rand(x.shape[1], 1, dtype = t.float64, requires_grad=True) # 30*1 weight matrix
    self.bias = t.rand(1, dtype = t.float64, requires_grad=True) # 1*1 bias matrix

  def forward(self, x):
    z = t.matmul(x, self.weights) + self.bias
    y_pred = t.sigmoid(z)
    return y_pred

 ## binary cross entropy loss for binary classification
  def loss_function(self, y_pred, y):
    epsilon = 1e-7
    y_pred = t.clamp(y_pred, epsilon, 1-epsilon)
    loss = -(y_train_tensor * t.log(y_pred) + (1-y_train_tensor) * t.log(1-y_pred)).mean()
    return loss

```

## defining params

```python
l_r = 0.01
epochs = 250
```

# training pipeline

```python
model = cancer_classifier(X_train_tensor)
```

```python
model.weights
```

```python
model.bias
```

# forward pass
# y_pred = w*x + b

```python
for epoch in range(epochs):
  y_pred = model.forward(X_train_tensor)
  loss = model.loss_function(y_pred, y_train_tensor)


  loss.backward()

  # parameter updation

  with t.no_grad():
    model.weights -=  l_r * model.weights.grad
    model.bias -=  l_r * model.bias.grad

  #zero gradients
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  print(f'epoch: {epoch+1}, loss: {loss.item()}')
```

```python
# model eval
with t.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred_class = (y_pred > 0.9).float()
```
